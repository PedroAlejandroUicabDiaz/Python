```python
# #Import matplotlib and numpy libraries
import numpy as np
import matplotlib.pyplot as plt

#Trigonometric function
def trigonometric_function(x):
    return np.sin(x)

# Values that X can take
x=np.linspace(0,10,100)

#Grafic from 0 to 10
plt.plot(x,trigonometric_function(x),color='r')

#Grafic's title
plt.title('Sinus')

plt.tight_layout()
plt.show()
```

```python
# Replica from Ing. Juan Vazquez Montejo's dataset
#Import matplotlib and pandas libraries
import matplotlib.pyplot as plt
import pandas as pd

#Reading of content from wtime.T001.his file and create a plot with matplotlib
data_time1=pd.read_csv('wktime.T001.his',delimiter=' ',header=None)
plt.plot(data_time1[0], data_time1[1], color='m', label='Time 1')

#Reading of content from wtime.T004.his file and create a plot with matplotlib
data_time4=pd.read_csv('wktime.T004.his',delimiter=' ',header=None)
plt.plot(data_time4[0], data_time4[1], color='g', label='Time 4')

#Graphic modification
#labels
plt.xlabel('Wealth')
plt.ylabel('Probability')
#Graphic's title
plt.title('Probability Distribution')
plt.legend()
#Scale type
plt.xscale("log")
plt.yscale("log")
#limits
plt.xlim((0.000001, 100.0))
plt.ylim((0.000001, 100000.0))
plt.tight_layout()
plt.show()
```
